## Review assignments

```python
annotations_11.active.value_counts()
```

```python
annotations_11 = annotations_11[['title', 'author', 'year', 'poem_id', 'book_id', 'id_poem', 
                                 'id_topic', 'value', 'timestamp', 'id_assignment', 'name']]
```

```python
annotations_11['book_poem_id'] = annotations_11['book_id'].astype(str) + '_' + annotations_11['poem_id']
```

```python
annotations_11.head()
```

### Counts

```python
s1 = annotations_11[annotations_11.value == 1].name.value_counts()
print(len(s1))
s1
```

```python
s2 = annotations_11[annotations_11.value == 0].name.value_counts()
s2
```

```python
missing_side_topics = (set(s1.index) - set(s2.index))

x = pd.Series([0]*len(missing_side_topics), index =[t for t in missing_side_topics])

s2 = s2.append(x)
s2
```

```python
# Creating a dictionary by passing Series objects as values
frame = {'Major topic': s1,
         'Minor topic': s2}
 
# Creating DataFrame by passing Dictionary
result = pd.DataFrame(frame)

result.sort_values(by=['Major topic'], ascending=False, inplace=True)
result
```

```python
import matplotlib.pyplot as plt
import numpy as np

plt.rcParams.update({'font.size': 18})

# plot data
x = np.arange(len(result.index))  # x locations of bars
width = 0.35  # width of bars

fig, ax = plt.subplots(figsize=(16,10))
rects1 = ax.bar(x - width/2, result["Major topic"], width, label="Major topic")
rects2 = ax.bar(x + width/2, result["Minor topic"], width, label="Minor topic")

# add some text for labels, title and custom x-axis tick labels, etc.
ax.set_ylabel("Count")
# ax.set_title("Major and Minor topics by topic name")
ax.set_xticks(x)
ax.set_xticklabels(result.index, rotation=45, ha='right')
ax.legend()

plt.savefig('annotated-data-distribution.png', dpi=400, bbox_inches='tight', pad_inches=0)

# display the plot
plt.show()
```

### Counts of major/minor topic assignment

```python
def create_histogram(data, filename):
    # Create the histogram
    fig, ax = plt.subplots(figsize=(8, 8))
    ax.bar(data.index, data, color='tab:blue')

    # Add labels and title
    ax.set_xlabel('Frequency')
    ax.set_ylabel('Count')
    
    ax.set_ylim([0, 400])

    # adjust the figure layout to make sure all x-axis labels are visible
    plt.xticks(data.index)
    
    plt.savefig(filename, dpi=400, bbox_inches='tight', pad_inches=0)  
    plt.show()
```
