# Store book_poem_id, lemmas and their POS tags

```python
df['book_poem_id'] = df['book_id'].astype(str) + '_' + df['poem_id']
```

```python
df = df[['book_poem_id', 'body']]
```

```python
lemmas_col = []

# extract lemmas and pos tags
for index, poem in df.iterrows():
    if index % 100 == 0:
        print(index, end=' ')
        
    lemmas_poem = []
    body = literal_eval(poem['body'])
    for stanza in body:
        for line in stanza:
            for word in line['words']:
                # if POS tag is Noun, Adjective or Verb
                pos = word['morph'][0]
                lemmas_poem.append((word['lemma'], pos))
    lemmas_col.append(lemmas_poem)
```

```python
df['lemmas'] = lemmas_col
```

```python
df = df[['book_poem_id', 'lemmas']]
```
